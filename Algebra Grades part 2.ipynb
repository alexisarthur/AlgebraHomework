```python
import pandas as pd

Location = "datasets/algebradata.csv"
df= pd.read_csv(Location)
```

```python
# Convert gender to numeric

def score_to_numeric(x):
    if x=='F':
        return 1
    if x=='M':
        return 0
```

```python
# Apply gender to numeric

df['Gender_val'] = df['Gender'].apply(score_to_numeric)
df.head()
```

```python
def score_to_numeric(x):
    if x=='A':
        return 5
    if x=='B':
        return 4
    if x=='C':
        return 3
    if x=='D':
        return 2
    if x=='E':
        return 1
    if x=='F':
        return 0
```

```python
df['Grade_val'] = df['Grade'].apply(score_to_numeric)
df.head()
```

```python
df.columns = ['First', 'Last', 'Gender','Grade','Hours','Gender_val','Grade_val']
df.head()
```

# Correlation & Regression Analysis

```python
# Correlation

df.head()

df.corr()
```

Strongest correlation is between hours of study and grade, though it is not very strong.

```python
# Regression 

import statsmodels.formula.api as sm
result = sm.ols(formula='Grade_val ~ Hours + Gender_val', data=df).fit()
result.summary()
```
